```python
import json

from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


with open('paths_config.json') as f:
    config = json.load(f)
    
evaluation_path = config['evaluation']

results = pd.read_csv(evaluation_path, sep=';')
display(results)
```

```python
results.sort_values(by='accuracy', inplace=True)
display(results)
```

```python
best_ts_classifiers_labels = {
    'baseline', 'ts+distknn(k=20)', 'ts+concatnn(act=relu,net=20,n=2400)', 'ts+linproj(c=50)'
}

best_ts_classifiers = results.loc[results['algorithm'].isin(best_ts_classifiers_labels)]

display(best_ts_classifiers)
```

```python
plt.figure();
fig, ax = plt.subplots();
ind = np.arange(len(best_ts_classifiers))
width = 0.35
plt.ylim((0.0, 1.0));
accuracy_rects = ax.bar(ind, best_ts_classifiers['accuracy'].tolist(), width, color='g');
overlap_rects = ax.bar(ind + width, best_ts_classifiers['average taxonomic overlap'].tolist(), width, 
                       color='b');
ax.set_xticks(ind+width);
xticklabels = list(map(lambda s: s.split('(')[0], best_ts_classifiers['algorithm'].tolist()))
ax.set_xticklabels(xticklabels);
ax.set_yticks(np.arange(0.0, 1.1, 0.1));
plt.grid();
plt.legend((accuracy_rects, overlap_rects), ('accuracy', 'average taxonomic overlap'), loc=0);

plt.savefig(config['best classifier evaluation plot'], bbox_inches='tight');
plt.show();
```
