```python
import json

from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


with open('paths_config.json') as f:
    config = json.load(f)
    
evaluation_path = config['evaluation']

results = pd.read_csv(evaluation_path, sep=';')
display(results)
```

```python
results.sort_values(by='accuracy', inplace=True, ascending=False)
display(results)
```

```python
best_ts_classifiers_labels = [
    'baseline', 'ts+distknn(k=20)', 'ts+concatnn(act=relu,net=20,h=3,n=1200)', 'ts+linproj(c=50)'
]

best_ts_classifiers = results.loc[results['algorithm'].isin(best_ts_classifiers_labels)]

display(best_ts_classifiers)
```

```python
labels = list(
    map(lambda l: l.split('+')[1] if '+' in l else l.split('+')[0], 
        map(lambda s: s.split('(')[0], 
            best_ts_classifiers['algorithm'].tolist())))

plt.figure();
fig, ax = plt.subplots();
ind = np.arange(len(best_ts_classifiers))
width = 0.35
plt.ylim((0.0, 1.0));
accuracy_rects = ax.bar(ind, best_ts_classifiers['accuracy'].tolist(), width, color='g');
overlap_rects = ax.bar(ind + width, best_ts_classifiers['average taxonomic overlap'].tolist(), width, 
                       color='b');
ax.set_xticks(ind+width);
ax.set_xticklabels(labels);
ax.set_yticks(np.arange(0.0, 1.1, 0.1));
ax.set_yticklabels(['{}%'.format(v) for v in np.arange(0, 101, 10)])
plt.grid();
plt.legend((accuracy_rects, overlap_rects), ('accuracy', 'average taxonomic overlap'), loc=0);
ax.set_title('Performance of classifiers')

plt.savefig(config['best classifier evaluation plot'], bbox_inches='tight');
plt.show();
```

```python
overlap_path = config['local taxonomic overlaps']

min_size = 10000
overlaps = dict()
for algorithm in best_ts_classifiers['algorithm'].tolist():
    with open(overlap_path.format(algorithm)) as f:
        overlaps[algorithm] = np.array(list(map(float, f.readline().strip().split(','))))
        if overlaps[algorithm].size < min_size:
            min_size = overlaps[algorithm].size

for algorithm in best_ts_classifiers['algorithm'].tolist():
    overlaps[algorithm] = overlaps[algorithm][:min_size]

plt.figure();
plt.ylim((0.0, 1.0));
width = 0.7 / len(best_ts_classifiers_labels)
x = np.vstack([overlaps[algorithm] 
               for algorithm in best_ts_classifiers['algorithm'].tolist()]).T
weights = np.ones(x.shape) / min_size
plt.hist(x, bins=10, label=labels, weights=weights);
plt.legend();
plt.grid();
plt.xlabel('local taxonomic overlap')
plt.ylabel('percentage of predictions')
plt.title('histogram of local taxonomic overlaps')
plt.savefig(config['best classifier overlap hist'], bbox_inches='tight');
plt.show();
```

```python
ks = [5, 10, 15, 20]
knn_classifiers_labels = ['ts+distknn(k={})'.format(k) for k in ks]

knn_classifiers = results.loc[results['algorithm'].isin(knn_classifiers_labels)]

display(knn_classifiers)
```

```python
plt.figure();
fig, ax = plt.subplots();
ind = np.arange(len(knn_classifiers))
width = 0.35
plt.ylim((0.0, 1.0));
accuracy_rects = ax.bar(ind, knn_classifiers['accuracy'].tolist(), width, color='g');
overlap_rects = ax.bar(ind + width, knn_classifiers['average taxonomic overlap'].tolist(), width, 
                       color='b');
ax.set_xticks(ind+width);
ax.set_xticklabels(list(map(str, ks)));
ax.set_yticks(np.arange(0.0, 1.1, 0.1));
ax.set_yticklabels(['{}%'.format(v) for v in np.arange(0, 101, 10)])
ax.set_xlabel('k neighbors')
plt.grid();
plt.legend((accuracy_rects, overlap_rects), ('accuracy', 'average taxonomic overlap'), loc=0);
ax.set_title('Performance of kNN classifiers')


plt.savefig(config['ts+distknn evaluation plot'], bbox_inches='tight');
plt.show();
```
